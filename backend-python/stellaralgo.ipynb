```python
import pandas as pd

# display full data set, otherwise it will only show the head and tail
# only do this when datasets are small
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# read in the file and display it
df = pd.read_parquet("./stellaralgo_dataset.parquet")
df
```

```python
# query/filter by date and sum up prices
total_price_per_event = df.query("`Purchase Date` == '02-Apr-23'").groupby('Event Name')['Price'].sum().reset_index(name='Total')
total_price_per_event
```

```python
ticket_count = df.groupby(['Event Name', 'Ticket Type']).size().reset_index(name='Ticket Count')
ticket_count
```

```python
# group by name and sum by price, then find the max
# you could one line this, but I wanted to check that I was doing it right
total_price_by_first_name = df.groupby('First Name')['Price'].sum()
richest_bitch = total_price_by_first_name.idxmax() 
```

```python
total_price_by_first_name
```

```python
richest_bitch
```

```python
# same thing but ticket count
# counting by first name because I didn't see another way
ticket_count_per_person = df['First Name'].value_counts()
fomo_bitch = ticket_count_per_person.idxmax()
```
